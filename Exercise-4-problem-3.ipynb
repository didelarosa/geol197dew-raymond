```python
temp_celsius = []
for i in temp_data:
    c = tf.fahr_to_celsius(i)
    temp_celsius.append(c)
    
temp_classes = []
for j in temp_celsius:
    temp_class = tf.temp_classifier(j)
    temp_classes.append(temp_class)
```

```python
zeros = temp_classes.count(0)
ones = temp_classes.count(1)
twos = temp_classes.count(2)
threes = temp_classes.count(3)
fours = temp_classes.count(4)
```
